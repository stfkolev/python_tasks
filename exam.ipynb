### First Exercise

```python
list1 = [10, 20, 4]
list2 = [70, 11, 20, 4, 100]

def getSecondMax(numbersList: list) -> int:
    maxNumber, currentNumber, secondMaxNumber = 0, 0, 0

    for number in numbersList:
        if number > maxNumber:
            secondMaxNumber = maxNumber
            maxNumber = number
        else:
            if number > currentNumber and currentNumber >= secondMaxNumber:
                secondMaxNumber = number
            
            currentNumber = number
    
    return secondMaxNumber

print(getSecondMax(list1) == 10)
print(getSecondMax(list2) == 70)
```

### Second Exercise

```python
class Auto:
    def __init__(self):
        self.model = ''
        self.color = ''
        self.productionYear = 1970
        self.engineType = ''
        self.enginePower = 0
        self.eurostandard = ''
        self.kilometers = 0
        self.extras = []

    ## Overrides
    def __repr__(self):
        return f'''Model: {self.model} -- Color: {self.color} -- Production Year: {self.productionYear} -- Engine Type: {self.engineType} -- Engine Power: {self.enginePower} -- EuroStandard: {self.eurostandard} -- Kilometers: {self.mileage} -- Extras: {[extra for extra in self.extras]}
        '''
    def __str__(self):
        return f'''Model: {self.model} -- Color: {self.color} -- Production Year: {self.productionYear} -- Engine Type: {self.engineType} -- Engine Power: {self.enginePower} -- EuroStandard: {self.eurostandard} -- Kilometers: {self.mileage} -- Extras: {[extra for extra in self.extras]}
        '''

    def __gt__(self, obj):
        return self.mileage > obj.mileage

    ## Static Methods
    @staticmethod
    def importVehicles():
        vehicles = []

        for _ in range(10):
            seeder = Faker()
            vehicle = Auto()

            vehicle.model = seeder.company()
            vehicle.color = seeder.color_name()
            vehicle.productionYear = seeder.random_int(1970, 2021)
            vehicle.engineType = 'petrol' if seeder.random_int(0, 1) else 'diesel'
            vehicle.enginePower = seeder.random_int(75, 1000)
            vehicle.eurostandard = f'EURO{str(seeder.random_int(1, 6))}'
            vehicle.mileage = seeder.random_int(10000, 500000)

            for _ in range(seeder.random_int(2, 8)):
                vehicle.extras.append(seeder.cryptocurrency_name())

            vehicles.append(vehicle)

        return vehicles

    @staticmethod
    def minAndMaxKilometers(vehicles: list):
        return min(vehicles, key=lambda vehicle: vehicle.kilometers), max(vehicles, key=lambda vehicle: vehicle.kilometers)

    @staticmethod
    def getAllVehiclesByModel(vehicles: list, model: str):
        return [vehicle for vehicle in vehicles if vehicle.model == model]

vehicles = Auto.importVehicles()

print('\n\nGenerated vehicles: \n\n')

for vehicle in vehicles:
    print(vehicle)

print('\n\nLowest and highest mileage vehicles: \n\n')

minMileageVehicle, maxMileageVehicle = Auto.minAndMaxKilometers(vehicles)
print(f'Minimum mileage vehicle: {minMileageVehicle}')
print(f'Maximum mileage vehicle: {maxMileageVehicle}')

print('\n\n Search for vehicle by model specified by the user: \n\n')
model = input('Enter vehicle model to search by: ')
print(Auto.getAllVehiclesByModel(vehicles, model))
```

### State

```python
import operator

class State:
    def __init__(self):
        self.name = ''
        self.language = ''
        self.capital = ''
        self.population = 0
        self.gdp = 0.0
        self.cities = []

    def populate(self) -> list:
        states = []

        for index in range(10):
            state = State()
            state.name = 'Bulgaria' + str(index)
            state.language = 'BG' + str(index ** 2)
            state.capital = 'BG' + str(index ** 2)
            state.population = index ** 4
            state.gdp = float(state.population ** 3)
            
            for cityIndex in range(3):
                state.cities.append('City' + str(cityIndex + index))

            states.append(state)
        
        return states
    
    def findStateByName(self, states, stateName):
        foundStates = []

        for state in states:
            if state.name == stateName:
                foundStates.append(state)

        return foundStates

    def statesWithBiggestPopulationAndGDP(self, states):
        return sorted(states, key= operator.attrgetter('population', 'gdp'), reverse=True)

stateObj = State()

print('\n\nAll States: \n\n')
states = stateObj.populate()

for state in states:
    print('State Name: ' + state.name + ' -- State GDP: ' + str(state.gdp) + ' -- State Population: ' + str(state.population))

print('\n\nFound states States: \n\n')
foundStates = stateObj.findStateByName(states, input('Enter state name'))

for foundState in foundStates:
    print('State Name: ' + foundState.name + ' -- State GDP: ' + str(foundState.gdp) + ' -- State Population: ' + str(foundState.population))


print('\n\nBiggest States: \n\n')
biggestStates = stateObj.statesWithBiggestPopulationAndGDP(states)

for biggestState in biggestStates:
    print('State Name: ' + biggestState.name + ' -- State GDP: ' + str(biggestState.gdp) + ' -- State Population: ' + str(biggestState.population))
```

```python
import operator

class Computer:
    def __init__(self):
        self.brand = ''
        self.model = ''
        self.cpuModel = ''
        self.frequency = ''
        self.memory = ''
        self.disk = ''
        self.price = 0.0

    def populate(self) -> list:
        computers = []

        for index in range(10):
            computer = Computer()
            computer.brand = f'Brand{index}' 
            computer.model = f'Model{index**2}'
            computer.cpuModel = f'Intel i{index}-{index**6}'
            computer.frequency = f'{index ** 1.5} Ghz'
            computer.memory = f'{index**2}GB'
            computer.disk = f'{index ** 4}GB'
            computer.price = float(index ** 2)
        
            computers.append(computer)
        
        return computers
    
    def findComputerByCPU(self, computers, cpuModel):
        foundComputers = []

        for computer in computers:
            if computer.cpuModel == cpuModel:
                foundComputers.append(computer)

        return foundComputers

    def lowestAndHighestPrice(self, computers):
        sortedComputers = sorted(computers, key= operator.attrgetter('price'), reverse=True)
        return [sortedComputers[0], sortedComputers[-1]]

computerObj = Computer()

print('\n\nAll States: \n\n')
computers = computerObj.populate()

for computer in computers:
    print(f'Computer Brand: {computer.brand} -- Computer Model:{computer.model} -- Computer CPU: {computer.cpuModel} -- CPU Frequency: {computer.frequency} -- Computer Memory: {computer.memory} -- Computer Disk: {computer.disk} -- Computer Price: {computer.price}')

print('\n\nFound computers: \n\n')
foundComputers = computerObj.findComputerByCPU(computers, input('Enter CPU Model'))

for foundComputer in foundComputers:
    print(f'Computer Brand: {foundComputer.brand} -- Computer Model:{foundComputer.model} -- Computer CPU: {foundComputer.cpuModel} -- CPU Frequency: {foundComputer.frequency} -- Computer Memory: {foundComputer.memory} -- Computer Disk: {foundComputer.disk} -- Computer Price: {foundComputer.price}')

print('\n\nLowest and Highest Price Computers: \n\n')
lowestAndHighestPriceComputers = computerObj.lowestAndHighestPrice(computers)

for computer in lowestAndHighestPriceComputers:
    print(f'Computer Brand: {computer.brand} -- Computer Model:{computer.model} -- Computer CPU: {computer.cpuModel} -- CPU Frequency: {computer.frequency} -- Computer Memory: {computer.memory} -- Computer Disk: {computer.disk} -- Computer Price: {computer.price}')
```

### Time Class

```python
import datetime

class Time:
    def __init__(self, hours: list):
        self.hours = hours
        self.time = datetime.timedelta()

    def addTime(self):
        for hour in self.hours:
            (hour, minute) = hour.split(':')
            deltaTime = datetime.timedelta(hours=int(hour), minutes=int(minute))

            self.time += deltaTime
        
        return self.time

    def displayTime(self):
        print(self.time)

    def displayMinutes(self, time):
        (hour, minute) = time.split(':')
        deltaTime = datetime.timedelta(hours=int(hour), minutes=int(minute))
        
        print(deltaTime.seconds // 60)


timestamps = ['00:00', '00:15', '09:30']
time = Time(timestamps)
time.addTime()

time.displayTime()
time.displayMinutes('18:43')
```
